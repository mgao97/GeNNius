```python
import matplotlib.pyplot as plt

# 
plt.rcParams['font.family'] = 'Arial'  # Arial
# plt.rcParams['font.weight'] = 'bold'   # normal-;bold-
plt.rcParams['font.size'] = 16

# 
x_values = [16, 32, 64, 128] # hidden size
y_values = [0.8816, 0.8914, 0.8970, 0.8985]

# # 
# plt.plot(x_values, y_values, marker='o', linestyle='-')
# 
plt.plot(range(len(x_values)), y_values, marker='o', linestyle='-')

# 
plt.title('DrugBank')
plt.xlabel('Different hidden size')
plt.ylabel('AUC')

#  x_values
plt.xticks(range(len(x_values)), x_values)

# 
plt.grid(True, linestyle='--')
plt.savefig('figs/drugbank_hidsize.pdf',bbox_inches='tight',pad_inches=0.1)

# 
plt.show()
```
